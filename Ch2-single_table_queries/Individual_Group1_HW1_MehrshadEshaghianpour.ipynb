- **Proposition:**
    
    - The proposition can be inferred from the `ROW_NUMBER()` function and by the use of `OVER (PARTITION BY custid ORDER BY orderdate, orderid) AS rownum`

- **Table:**
    
    - The table involved is <span style="font-size: 12px;"><code>Sales.Orders</code></span>

- **Columns:**
    
    - Columns selected in the query are `custid, orderdate`, and <span style="color: rgb(163, 21, 21); font-family: Menlo, Monaco, Consolas, &quot;Droid Sans Mono&quot;, &quot;Courier New&quot;, monospace, &quot;Droid Sans Fallback&quot;; font-size: 12px;">orderid</span>

```sql
USE TSQLV4

SELECT
	custid,
	orderdate,
    orderid,
    ROW_NUMBER() OVER (PARTITION BY custid ORDER BY orderdate, orderid) AS rownum
FROM
	Sales.Orders
ORDER BY
	custid,
    rownum;

USE Northwinds2022TSQLV7

SELECT
	CustomerId,
	OrderDate,
    Orderid,
    ROW_NUMBER() OVER (PARTITION BY CustomerId ORDER BY OrderDate, Orderid) AS rownum
FROM
	Sales.[Order]
ORDER BY
	CustomerId,
    rownum;
```

- **Proposition:**
    
    - The proposition is in the `WHERE` clause
        
    - `region = N'WA'` returns customers located in the WA region
        
    - `region <> N'WA'` returns customers located in regions other than WA
        
    - `region = NULL` returns nothing because it is not a valid comparison
        
    - `region IS NULL` returns customers whose region value is `NULL`
        
    - `region <> N'WA' OR region IS NULL` returns customers located in regions other than WA including those regions whose value is `NULL`
        
- **Table:**
    
    - The table involved is <span style="font-size: 12px;"><code>Sales.Customers</code></span>

- **Columns:**
    
    - Columns selected in the query are `custid, country, region,` and `city`
        
- **Predicate:**
    
    - The predicate here is in the `WHERE` clause
        
    - `region = N'WA'` 
        
    - `region <> N'WA'` 
        
    - `region = NULL` 
        
    - `region IS NULL` 
        
    - `region <> N'WA' OR region IS NULL` 

```sql
USE TSQLV4;

SELECT custid, country, region, city
FROM Sales.Customers
WHERE region = N'WA';

SELECT custid, country, region, city
FROM Sales.Customers
WHERE region <> N'WA';

SELECT custid, country, region, city
FROM Sales.Customers
WHERE region = NULL;

SELECT custid, country, region, city
FROM Sales.Customers
WHERE region IS NULL;

SELECT custid, country, region, city
FROM Sales.Customers
WHERE region <> N'WA'
   OR region IS NULL;

USE Northwinds2022TSQLV7;

SELECT CustomerId, CustomerCountry, CustomerRegion, CustomerCity
FROM Sales.Customer
WHERE CustomerRegion = N'WA';

SELECT CustomerId, CustomerCountry, CustomerRegion, CustomerCity
FROM Sales.Customer
WHERE CustomerRegion <> N'WA';

SELECT CustomerId, CustomerCountry, CustomerRegion, CustomerCity
FROM Sales.Customer
WHERE CustomerRegion = NULL;

SELECT CustomerId, CustomerCountry, CustomerRegion, CustomerCity
FROM Sales.Customer
WHERE CustomerRegion IS NULL;

SELECT CustomerId, CustomerCountry, CustomerRegion, CustomerCity
FROM Sales.Customer
WHERE CustomerRegion <> N'WA'
   OR CustomerRegion IS NULL;
```
