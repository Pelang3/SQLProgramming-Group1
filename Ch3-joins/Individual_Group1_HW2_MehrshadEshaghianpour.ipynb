- **Proposition:**
    
    - The proposition in this query is expressed in the `SELECT` clause
        
    - `C.custid` and `C.companyname`
        
    - For each customer, whether they placed an order on '20220212'
        
    - It also includes customers who have not placed any orders
        
    - `O.orderid IS NULL`
        
- **Table:**
    
    - `Sales.Customers` aliased as `C`
        
    - `Sales.Orders` aliased as `O`
        
- **Columns:**
    
    - Columns selected in the query are `C.custid` and `C.companyname` 
        
- **Predicate:**
    
    - `O.orderdate = '20220212'` filters the rows from the `Sales.Orders` table to include only those orders that were placed on the date '20220212'

```sql
USE TSQLV4

SELECT DISTINCT C.custid, C.companyname,
    CASE WHEN O.orderid IS NOT NULL THEN 'Yes' ELSE 'No' END AS HasOrderOn20220212
FROM Sales.Customers AS C
    LEFT OUTER JOIN Sales.Orders AS O
    ON O.custid = C.custid
        AND O.orderdate = '20220212';

USE Northwinds2022TSQLV7

SELECT DISTINCT C.CustomerId, C.CustomerCompanyName,
    CASE WHEN O.orderid IS NOT NULL THEN 'Yes' ELSE 'No' END AS HasOrderOn20220212
FROM Sales.Customer AS C
    LEFT OUTER JOIN Sales.[Order] AS O
    ON O.CustomerId = C.CustomerId
        AND O.orderdate = '20220212';
```

<span style="color: rgb(0, 128, 0);">-- Multi-Join Queries</span>

<span style="color: rgb(0, 128, 0);">-----------------------</span>

- **Proposition:**
    
    - The proposition in this query is expressed in the `SELECT` clause
        
    - `C.custid, C.companyname, O.orderid, OD.productid,` and `OD.qty`
        
    - Specifies the columns that are to be used

    - `JOIN` indicate how the tables are related and should be combined
        
- **Table:**
    
    - `Sales.Customers` aliased as `C`
        
    - `Sales.Orders` aliased as `O` 
    
    - `Sales.OrderDetails` as `OD`
        
- **Columns:**
    
    - Columns selected in the query are `C.custid, C.companyname, O.orderid, OD.productid,` and `OD.qty`
        
- **Predicate:**
    
    - `INNER JOIN Sales.Orders AS O ON C.custid = O.custid` the `custid` column in the Customers table must be equal to the `custid` column in the `Orders` table for a row to be included in the result

    - `INNER JOIN Sales.OrderDetails AS OD ON O.orderid = OD.orderid` the `orderid` column in the Orders table must be equal to the `orderid` column in the `OrderDetails` table for a row to be included in the result

```sql
USE TSQLV4

SELECT
    C.custid, C.companyname, O.orderid,
    OD.productid, OD.qty
FROM Sales.Customers AS C
    INNER JOIN Sales.Orders AS O
    ON C.custid = O.custid
    INNER JOIN Sales.OrderDetails AS OD
    ON O.orderid = OD.orderid;

USE Northwinds2022TSQLV7

SELECT
    C.CustomerId, C.CustomerCompanyName, O.orderid,
    OD.ProductId, OD.Quantity
FROM Sales.Customer AS C
    INNER JOIN Sales.[Order] AS O
    ON C.CustomerId = O.CustomerId
    INNER JOIN Sales.OrderDetail AS OD
    ON O.orderid = OD.orderid;
```

<span style="color: rgb(0, 128, 0);">-- Fundamentals of Outer Joins </span>

<span style="color: rgb(0, 128, 0);">-----------------------</span>

- **Proposition:**
    
    - The proposition is implied by the condition specified in the `ON` clause of the `LEFT OUTER JOIN`
        
    - For each customer `C.custid` there may or may not exist corresponding `O.orderid` and if there are orders, they will be joined based on the condition that the `custid` values match between the two tables
        
- **Table:**
    
    - `Sales.Customers` aliased as `C`
        
    - `Sales.Orders` aliased as `O` 
        
- **Columns:**
    
    - Columns selected in the query are `C.custid, C.companyname, O.orderid,` and `OD.productid`
        
- **Predicate:**
    
    - `ON C.custid = O.custid` indicates that the query will return rows where the `custid` column in the Customers table matches the `custid` column in the Orders table.

```sql
-- Customers and their orders, including customers with no orders
USE TSQLV4

SELECT C.custid, C.companyname, O.orderid
FROM Sales.Customers AS C
  LEFT OUTER JOIN Sales.Orders AS O
    ON C.custid = O.custid;

USE Northwinds2022TSQLV7

SELECT C.CustomerId, C.CustomerCompanyName, O.OrderId
FROM Sales.Customer AS C
  LEFT OUTER JOIN Sales.[Order] AS O
    ON C.CustomerId = O.CustomerId

-- Customers with no orders
USE TSQLV4

SELECT C.custid, C.companyname
FROM Sales.Customers AS C
  LEFT OUTER JOIN Sales.Orders AS O
    ON C.custid = O.custid
WHERE O.orderid IS NULL;

USE Northwinds2022TSQLV7

SELECT C.CustomerId, C.CustomerCompanyName
FROM Sales.Customer AS C
  LEFT OUTER JOIN Sales.[Order] AS O
    ON C.CustomerId = O.CustomerId
WHERE O.orderid IS NULL;
```
